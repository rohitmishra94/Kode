```python
from itertools import combinations

tech = ['pandas', 'tensorflow', 'pytorch', 'numpy', 'python']
keywords = ['machine learning']

# Generate unique combinations of 2 elements from the 'tech' list
combinations_list = list(combinations(tech, 2))

# Combine each combination with the 'keywords' list
result_list = [combo + tuple(keywords) for combo in combinations_list]

# Display the result
for result in result_list:
    print(result)
```

```python
import requests
import time 
url = "https://api.github.com/search/users"

headers = {
    "Authorization": "Bearer git_token"
}
all_list = []
tech = list(map(lambda x: x[0]+' '+x[1]+' '+x[2],result_list))
# for i in range(3):
for t in tech:
    print(t)
    for j in range(10):
        print('page ',j)
        params = {
        "q": f"language:{t} repos:>10 sort:forks",
        "per_page": 100,
        "page": j}
        response = requests.get(url, params=params)#, headers=headers)
        data = response.json()
        if 'items' in data:
            print('working for page ',j)
            for k in data['items']:
                all_list.append((k['id'],k['login'],k['type'],k['url']))
        else:
            pos = int(response.headers['X-RateLimit-Reset'])-int(time.time())
            print(response.headers['X-RateLimit-Remaining'],pos)
            print('sleeping')
            time.sleep(61)

            

# Process the response here (e.g., print(response.json()))
# print(f'reset in {int(response.headers['X-RateLimit-Reset'])-1703283913}')
```
